DATA SPLITING

```python
from sklearn.model_selection import train_test_split

# Split the data into training, validation, and test sets
train_data, val_test_data = train_test_split(data, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(val_test_data, test_size=0.3, random_state=42)

print("Training data shape:", train_data.shape)
print("Validation data shape:", val_data.shape)
print("Test data shape:", test_data.shape)
```

```python
missing_values=train_data.isnull().sum()
# Sort the columns based on the number of missing values in descending order
sorted_missing_values = missing_values.sort_values(ascending=False)

# Select the top 20 columns with the highest number of missing values
top_20_missing_values = sorted_missing_values.tail(70)

# Display the result
print(top_20_missing_values)
```

Missing values replaced with mean

```python
for column in train_data.select_dtypes(include=['int', 'float']).columns:
    train_data[column].fillna((train_data[column].mean()), inplace=True)
```

```python
train_data['icu_death'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(train_data.drop('icu_death', axis=1), train_data['icu_death'])

```

```python
y_res.value_counts()
```

```python
from pycaret.classification import *
clf = setup(data = X_res, target = y_res, use_gpu=True)
best_model = compare_models()
tuned_model = tune_model(best_model)
final_model = finalize_model(tuned_model)

```
